***What is World Happiness Report (2021) ?***


The United Nationâ€™s Sustainable Development Solutions Network published the world happiness report on 19th March 2021 that ranks the world's 149 countries on how happy their citizens perceive themselves to be based on various indicators. The happiness study ranks the countries on the basis of questions from the Gallup World Poll. The results are then equated with other factors such as GDP, life expectancy, generosity, etc. In the year 2021, it focused on the effects of the Covid-19 pandemic and how people all over the world have managed to survive and prosper.

***What is GDP per Capita ?***
GDP stands for Gross Domestic Product. GDP per capita is a measure that calculates the economic output of a country divided by its total population. It is used to measure a country's economic productivity on a per-person basis and is often used as an indicator of a country's standard of living. GDP per capita is calculated by dividing a country's GDP by its total population, and it provides a useful way to compare the economic output of different countries, regardless of their size or population.

***What is Happiness Score ?***
 The Happiness Score is a measure of well-being and happiness of a country's citizens, based on several factors such as GDP per capita, social support, life expectancy, freedom to make life choices, generosity, and perceptions of corruption. The Happiness Score is a value between 0 and 10, with higher scores indicating greater levels of happiness and well-being.

```python
# plot b/w 'Happiness Score' and 'GDP per capita'
```

```python
plt.title("plot b/w 'Happiness Score' and 'Logged GDP per capita'")
sns.set(rc={'figure.figsize':(15,7)})
sns.scatterplot(x= data.Happiness_Score, y= data.Logged_GDP_per_capita, hue= data.Regional_Indicator, s=200)
plt.legend(loc='upper left',fontsize='10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita')
plt.show()
```

```python
# Correlation Map
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

cor = data.corr(method='pearson')
f, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(cor, mask=np.zeros_like(cor, dtype=bool), cmap='Blues', square=True, ax=ax)
```

***What is Perceptions of Corruption ?***
 "Perceptions_of_corruption" is a column that indicates the perceived levels of public sector corruption in each country, as reported in the Gallup World Poll. The values range from 0 to 1, where 0 indicates a perception of high levels of corruption, and 1 indicates a perception of low levels of corruption.

***What is Regional Indicator ?*** "Regional_Indicator" is a categorical variable that indicates the region to which each country belongs. The regions are:

Western Europe, North America and ANZ, Middle East and North Africa , Latin America and Caribbean, Central and Eastern Europe East Asia, Southeast Asia, Commonwealth of Independent States, Sub-Saharan Africa, South Asia.

```python
# Show the average corruption perception score for each region.

corruptions = data.groupby('Regional_Indicator')[['Perceptions_of_corruption']].mean()
corruptions
```

```python
# plot a bar chart of the mean perceptions of corruption in each region

plt.rcParams['figure.figsize'] = (12,8)
plt.title('Perception of the Corruption in various Regions')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Corruption', fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.bar(corruptions.index, corruptions.Perceptions_of_corruption)
```

***What is Countries Life Expectancy ?***
 "Healthy Life Expectancy" represents the life expectancy in years for each country.

```python
top_10 = data.head(10)
bottom_10 = data.tail(10)
```

```python
#  Plot for top 10 happiest countries based on healty life expectancy
fig, axes= plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.Country_Name
axes[0].set_title("Top 10 Happiest Countries Life Expectancy")
axes[0].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x=top_10.Country_Name, y=top_10.Healthy_Life_Expectancy, ax=axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')

#  Plot for top 10 Least happiest countries based on life expectancy
xlabels=bottom_10.Country_Name
axes[1].set_title("Top 10 Least Happiest Countries Life Expectancy")
axes[1].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x=bottom_10.Country_Name, y=bottom_10.Healthy_Life_Expectancy, ax=axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')
```

***What is the freedom to make life choices ?***

The "Freedom to make life choices" is one of the variables in the World Happiness Report dataset, which measures the national average score on the question "Are you satisfied or dissatisfied with your freedom to choose what you do with your life?". It is a measure of the perceived level of individual freedom in making important life decisions.

```python
# show the relationship between the freedom to make life choices and happiness score for different regions
plt.rcParams['figure.figsize'] = (15,7)
plt.title('Relationship between the freedom to make life choices and happiness score for different regions')
sns.scatterplot(x= data.Freedom_to_make_life_choices, y= data.Happiness_Score, hue= data.Regional_Indicator)
plt.legend(loc='upper left',fontsize='12')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness Score')
```
